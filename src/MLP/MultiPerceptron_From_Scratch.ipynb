```python
import numpy as np
import random
import _pickle as cPickle
import gzip
import os
import sys
import matplotlib.pyplot as plt
```

```python
from mlp import *
from mlp_utils import *
```

```python
(X_train, y_train), (X_test, y_test) = load_mnist("Datasets/mnist", num_training=9000, num_test=1000, cnn=False, one_hot=True)
batch_logs = {}
```

```python
batch_size=1
multi_perceptron = MultiPerceptron(input_size=X_train.shape[-1], output_size=y_train.shape[-1], hidden_structure=[20, 20], cost_function=MeanSquareRootCost, hidden_activation=SigmoidActivation)
log = multi_perceptron.sgd(X_train, y_train, batch_size, epochs=batch_size, alpha=1, verbose=True, X_test=X_test, y_test=y_test)
batch_logs["batch_{}".format(batch_size)] = log
plt.plot(batch_logs["batch_1"], label="batch_1_test_acc")
plt.legend()
plt.show()
```

```python
for i, batch_size in enumerate([1, 2, 4, 8, 16, 32, 64, 128, 256, 512, 1024, 2048]):
    logs = batch_logs["batch_{}".format(batch_size)]
    plt.subplot(6, 2, i+1)
    plt.plot(logs)
    plt.title(str(batch_size), pad=0)
    
plt.savefig("batch_size_test_acc.svg", format="svg", dpi=10000)
plt.show()
```
